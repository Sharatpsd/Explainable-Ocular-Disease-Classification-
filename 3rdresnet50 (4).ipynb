```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras import backend as K
import os
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================================================================
# 0. CONFIGURATION (REQUIRED TO EDIT)
# ==============================================================================
# !!! CRITICAL: Ensure these paths are correct after mounting Google Drive !!!
IMAGE_DIR = "/content/drive/MyDrive/thesis data/ODIR-5K/ODIR-5K/Training Images"
CSV_PATH = "/content/drive/MyDrive/thesis data/full_df.csv"
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
LEARNING_RATE_PHASE1 = 1e-3
LEARNING_RATE_PHASE2 = 1e-5
EPOCHS_PHASE1 = 5
EPOCHS_PHASE2 = 15
LABEL_COLUMNS = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
NUM_CLASSES = len(LABEL_COLUMNS)

# Define saving locations (using HDF5 format)
MODEL_SAVE_DIR = "model_weights"
os.makedirs(MODEL_SAVE_DIR, exist_ok=True)
WEIGHTS_PHASE1 = os.path.join(MODEL_SAVE_DIR, 'model_phase1_best.h5')
WEIGHTS_PHASE2 = os.path.join(MODEL_SAVE_DIR, 'model_phase2_best.h5')
# ==============================================================================

# ==============================================================================
# 1. DATA LOADING, SPLITTING, AND WEIGHT CALCULATION
# ==============================================================================
print("1. Loading Data, Splitting, and Calculating Weights...")
df = pd.read_csv(CSV_PATH)
df_model = df[['filename'] + LABEL_COLUMNS].copy()
train_val_df, test_df = train_test_split(df_model, test_size=0.15, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=(0.15 / 0.85), random_state=42)

class_weights = {}
train_labels = train_df[LABEL_COLUMNS].values
for i, label in enumerate(LABEL_COLUMNS):
    weights = compute_class_weight('balanced', classes=np.array([0, 1]), y=train_labels[:, i])
    class_weights[i] = weights[1]

# ==============================================================================
# 2. CUSTOM WEIGHTED LOSS FUNCTION (Required for HDF5 saving/loading)
# ==============================================================================
def weighted_binary_cross_entropy(y_true, y_pred):
    y_pred = K.clip(y_pred, K.epsilon(), 1 - K.epsilon())
    weight_vector = tf.constant([class_weights[i] for i in range(NUM_CLASSES)], dtype=tf.float32)
    pos_loss = -weight_vector * y_true * K.log(y_pred)
    neg_loss = -(1 - y_true) * K.log(1 - y_pred)
    return K.mean(pos_loss + neg_loss)

# ==============================================================================
# 3. DATA GENERATORS (Only printing the setup flow_from_dataframe results)
# ==============================================================================
print("\n2. Setting up Data Generators...")
train_datagen = ImageDataGenerator(rescale=1./255, rotation_range=20, horizontal_flip=True)
val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df, directory=IMAGE_DIR, x_col="filename", y_col=LABEL_COLUMNS,
    target_size=IMAGE_SIZE, batch_size=BATCH_SIZE, class_mode="raw", shuffle=True
)
validation_generator = val_test_datagen.flow_from_dataframe(
    dataframe=val_df, directory=IMAGE_DIR, x_col="filename", y_col=LABEL_COLUMNS,
    target_size=IMAGE_SIZE, batch_size=BATCH_SIZE, class_mode="raw", shuffle=False
)
test_generator = val_test_datagen.flow_from_dataframe(
    dataframe=test_df, directory=IMAGE_DIR, x_col="filename", y_col=LABEL_COLUMNS,
    target_size=IMAGE_SIZE, batch_size=BATCH_SIZE, class_mode="raw", shuffle=False
)

# ==============================================================================
# 4. MODEL ARCHITECTURE & TRAINING SETUP
# ==============================================================================
def create_model(num_classes):
    print("\n3. Creating Model Architecture (ResNet50)...")
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3), name='resnet50')
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(num_classes, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    return model, base_model

model, base_model = create_model(NUM_CLASSES)

# --- Phase 1: Compile and Train ---
print("\n=======================================================")
print("4. Starting Phase 1: Training Classification Head")
print("=======================================================")
for layer in base_model.layers: layer.trainable = False
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE_PHASE1), loss=weighted_binary_cross_entropy,
    metrics=[tf.keras.metrics.AUC(multi_label=True, name='auc'), tf.keras.metrics.BinaryAccuracy(name='binary_accuracy')]
)
checkpoint_phase1 = ModelCheckpoint(WEIGHTS_PHASE1, monitor='val_auc', mode='max', save_best_only=True, verbose=1)
model.fit(
    train_generator, epochs=EPOCHS_PHASE1, validation_data=validation_generator, callbacks=[checkpoint_phase1]
)

# --- Phase 2: Compile and Fine-Tune ---
print("\n=======================================================")
print("5. Starting Phase 2: Fine-Tuning Entire Network")
print("=======================================================")
if os.path.exists(WEIGHTS_PHASE1): model.load_weights(WEIGHTS_PHASE1)
for layer in base_model.layers: layer.trainable = True
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE_PHASE2), loss=weighted_binary_cross_entropy,
    metrics=[tf.keras.metrics.AUC(multi_label=True, name='auc'), tf.keras.metrics.BinaryAccuracy(name='binary_accuracy')]
)
checkpoint_phase2 = ModelCheckpoint(WEIGHTS_PHASE2, monitor='val_auc', mode='max', save_best_only=True, verbose=1)
model.fit(
    train_generator, epochs=EPOCHS_PHASE2, validation_data=validation_generator, callbacks=[checkpoint_phase2]
)

# ==============================================================================
# 5. FINAL EVALUATION AND VISUALIZATION
# ==============================================================================
print("\n=======================================================")
print("6. FINAL EVALUATION & PLOT GENERATION")
print("=======================================================")

# Load the final best model
if os.path.exists(WEIGHTS_PHASE2):
    # Use load_model to handle the custom loss function
    final_model = load_model(WEIGHTS_PHASE2, custom_objects={"weighted_binary_cross_entropy": weighted_binary_cross_entropy})
    print(f"Loaded final model from {WEIGHTS_PHASE2}")
else:
    print("FATAL ERROR: Final weights file not found. Cannot proceed to evaluation.")
    exit()

# Run Evaluation
results = final_model.evaluate(test_generator, verbose=1)
loss, auc, acc = results[0], results[1], results[2]

# Run Predictions
test_probabilities = final_model.predict(test_generator)
test_true_labels = test_generator.labels
test_predicted_labels = (test_probabilities > 0.5).astype(int)

# --- VISUALIZATION 1: CLASSIFICATION MATRIX (Metrics Summary) ---
print("\n--- Summary Test Set Results ---")
print(f"Test Loss: {loss:.4f}, Test AUC: {auc:.4f}, Test Binary Accuracy: {acc:.4f}")

report_dict = classification_report(test_true_labels, test_predicted_labels, target_names=LABEL_COLUMNS, output_dict=True, zero_division=0)
report_df = pd.DataFrame(report_dict).transpose().round(4).iloc[:-3]

plt.figure(figsize=(10, 6))
sns.heatmap(report_df[['precision', 'recall', 'f1-score']], annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5, linecolor='black')
plt.title('Multi-Label Classification Matrix (Precision, Recall, F1-Score)')
plt.ylabel('Disease Label')
plt.yticks(rotation=0)
plt.xlabel(f'Mean Test Binary Accuracy: {acc:.4f} | Test AUC: {auc:.4f}', fontsize=12)
plt.savefig('classification_matrix.png')
print("Visualization 1: Classification Matrix saved as classification_matrix.png")

# --- VISUALIZATION 2: CONFUSION MATRIX (Specific Disease: Diabetic Retinopathy 'D') ---
D_index = LABEL_COLUMNS.index('D')
D_true = test_true_labels[:, D_index]
D_pred = test_predicted_labels[:, D_index]
cm = confusion_matrix(D_true, D_pred)

plt.figure(figsize=(5, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['Actual Negative', 'Actual Positive'])
plt.title(f'Confusion Matrix for {LABEL_COLUMNS[D_index]} (Diabetic Retinopathy)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.savefig('confusion_matrix_D.png')
print("Visualization 2: Confusion Matrix for 'D' saved as confusion_matrix_D.png")

print("\n-----------------------------------------------------------")
print("ALL TRAINING AND VISUALIZATION CODE EXECUTED. Check your Colab environment for the plots.")
```

```python
# ==========================================================
# FULL THESIS PIPELINE – SAFE + UPDATED
# ==========================================================

# ================== IMPORTS ==================
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint

# ================== PATHS ==================
IMAGE_DIR = "/content/drive/MyDrive/thesis data/ODIR-5K/ODIR-5K/Training Images"
CSV_PATH  = "/content/drive/MyDrive/thesis data/full_df.csv"
SAVE_DIR  = "/content/drive/MyDrive/thesis_results"
os.makedirs(SAVE_DIR, exist_ok=True)

# ================== CONFIG ==================
IMAGE_SIZE = (224,224)
BATCH_SIZE = 32
EPOCHS = 10
LABEL_COLUMNS = ['N','D','G','C','A','H','M','O']
NUM_CLASSES = len(LABEL_COLUMNS)

# ==========================================================
# 1️⃣ LOAD DATA
# ==========================================================
df = pd.read_csv(CSV_PATH)
df = df[['filename'] + LABEL_COLUMNS + ['Left-Fundus','Right-Fundus']]

# ==========================================================
# 2️⃣ OPTION–1: PSEUDO EXTERNAL VALIDATION
#    Train → LEFT eye, Test → RIGHT eye
# ==========================================================
train_df = df[df['Left-Fundus'].notna()]
test_df  = df[df['Right-Fundus'].notna()]

train_df = train_df[['filename'] + LABEL_COLUMNS]
test_df  = test_df[['filename'] + LABEL_COLUMNS]

print("Train images:", len(train_df))
print("External test images:", len(test_df))

# ==========================================================
# 3️⃣ DATA GENERATORS
# ==========================================================
datagen = ImageDataGenerator(rescale=1./255)

train_gen = datagen.flow_from_dataframe(
    train_df, IMAGE_DIR, "filename", LABEL_COLUMNS,
    target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode="raw", shuffle=True
)

test_gen = datagen.flow_from_dataframe(
    test_df, IMAGE_DIR, "filename", LABEL_COLUMNS,
    target_size=IMAGE_SIZE, batch_size=BATCH_SIZE,
    class_mode="raw", shuffle=False
)

# ==========================================================
# 4️⃣ MODEL (MAIN MODEL)
# ==========================================================
base_model = ResNet50(weights="imagenet", include_top=False,
                      input_shape=(224,224,3))

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(512, activation="relu")(x)
x = Dropout(0.5)(x)
output = Dense(NUM_CLASSES, activation="sigmoid")(x)

model = Model(base_model.input, output)

for layer in base_model.layers:
    layer.trainable = False

model.compile(
    optimizer=Adam(1e-3),
    loss="binary_crossentropy",
    metrics=[tf.keras.metrics.AUC(multi_label=True), "binary_accuracy"]
)

checkpoint = ModelCheckpoint(
    SAVE_DIR + "/best_model.keras",
    monitor="auc",
    save_best_only=True,
    verbose=1
)

# ==========================================================
# 5️⃣ TRAIN
# ==========================================================
model.fit(
    train_gen,
    epochs=EPOCHS,
    callbacks=[checkpoint]
)

# ==========================================================
# 6️⃣ EVALUATION (EXTERNAL STYLE)
# ==========================================================
results = model.evaluate(test_gen)
print("External Test AUC:", results[1])
print("External Test Accuracy:", results[2])

# ==========================================================
# 7️⃣ CONFUSION MATRIX (Detailed)
# ==========================================================
preds = model.predict(test_gen)
y_true = test_gen.labels
y_pred = (preds > 0.5).astype(int)

report = classification_report(
    y_true, y_pred, target_names=LABEL_COLUMNS, output_dict=True
)
df_report = pd.DataFrame(report).transpose()

plt.figure(figsize=(10,6))
sns.heatmap(df_report[['precision','recall','f1-score']],
            annot=True, cmap="YlGnBu")
plt.title("Detailed Classification Metrics")
plt.savefig(SAVE_DIR + "/metrics_matrix.png")
plt.close()

# Single disease confusion (Diabetic Retinopathy)
idx = LABEL_COLUMNS.index('D')
cm = confusion_matrix(y_true[:,idx], y_pred[:,idx])

plt.figure(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix – Diabetic Retinopathy")
plt.savefig(SAVE_DIR + "/confusion_DR.png")
plt.close()

# ==========================================================
# 8️⃣ ABLATION STUDY (BASELINE – NO DROPOUT)
# ==========================================================
x2 = GlobalAveragePooling2D()(base_model.output)
x2 = Dense(512, activation="relu")(x2)
out2 = Dense(NUM_CLASSES, activation="sigmoid")(x2)

baseline_model = Model(base_model.input, out2)

baseline_model.compile(
    optimizer=Adam(1e-3),
    loss="binary_crossentropy",
    metrics=[tf.keras.metrics.AUC(multi_label=True)]
)

baseline_model.fit(train_gen, epochs=3)
base_results = baseline_model.evaluate(test_gen)

print("Baseline AUC (Ablation):", base_results[1])

# ==========================================================
# 9️⃣ SAVE EVERYTHING
# ==========================================================
model.save(SAVE_DIR + "/final_model.keras")
np.save(SAVE_DIR + "/preds.npy", preds)
np.save(SAVE_DIR + "/labels.npy", y_true)

print("✅ ALL DONE. Everything saved safely to Google Drive.")
```

```python
# ==========================================================
# HIGH-ACCURACY + SAFE SPLIT (PATIENT-WISE)
# ==========================================================

import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import GroupShuffleSplit
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# ---------------- CONFIG ----------------
IMAGE_DIR = "/content/drive/MyDrive/thesis data/ODIR-5K/ODIR-5K/Training Images"
CSV_PATH  = "/content/drive/MyDrive/thesis data/full_df.csv"

IMAGE_SIZE = (224,224)
BATCH_SIZE = 32
EPOCHS = 10

LABEL_COLUMNS = ['N','D','G','C','A','H','M','O']

# ---------------- LOAD DATA ----------------
df = pd.read_csv(CSV_PATH)

# Keep patient ID
df = df[['ID','filename'] + LABEL_COLUMNS]

# ---------------- PATIENT-WISE SPLIT ----------------
gss = GroupShuffleSplit(test_size=0.15, random_state=42)

train_idx, test_idx = next(gss.split(df, groups=df['ID']))

train_df = df.iloc[train_idx]
test_df  = df.iloc[test_idx]

print("Train patients:", train_df['ID'].nunique())
print("Test patients :", test_df['ID'].nunique())

# ---------------- GENERATORS ----------------
datagen = ImageDataGenerator(rescale=1./255)

train_gen = datagen.flow_from_dataframe(
    train_df,
    IMAGE_DIR,
    x_col="filename",
    y_col=LABEL_COLUMNS,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="raw",
    shuffle=True
)

test_gen = datagen.flow_from_dataframe(
    test_df,
    IMAGE_DIR,
    x_col="filename",
    y_col=LABEL_COLUMNS,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="raw",
    shuffle=False
)

# ---------------- MODEL ----------------
base_model = ResNet50(weights="imagenet", include_top=False,
                      input_shape=(224,224,3))

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(512, activation="relu")(x)
x = Dropout(0.5)(x)
output = Dense(len(LABEL_COLUMNS), activation="sigmoid")(x)

model = Model(base_model.input, output)

# Partial fine-tune (accuracy boost)
for layer in base_model.layers[:-40]:
    layer.trainable = False
for layer in base_model.layers[-40:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(1e-5),
    loss="binary_crossentropy",
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name="accuracy"),
        tf.keras.metrics.AUC(multi_label=True, name="auc")
    ]
)

# ---------------- TRAIN ----------------
model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=test_gen
)

# ---------------- EVALUATE ----------------
results = model.evaluate(test_gen)
print("FINAL TEST ACCURACY:", results[1])
print("FINAL TEST AUC:", results[2])
```
